## 前言：什么是编程？学编程对孩子有什么好处？

电脑编程，或者是写代码，对孩子来说，是一项关键技能，每个孩子都值得去学习。

在我们日常生活中，我们通过电脑解决问题，玩游戏，帮助我们更高效的工作,完成重复的人物，存储信息，创作新的东西，和全世界的朋友联系在一起。

我们每个人都可以学会编程，他就像解决一个拼图游戏或者一个谜语。现在就是学编程最好的时候。

我们处在史无前例的一个时代：以前从来没有人像我们现在这样，可以通过电脑与全世界数十亿的人进行链接。

我们生活在一个具有更多新的可能性的世界，从电动汽车，到无人驾驶，机器护理人员，甚至机器人可以送快递，送外卖。

如果你的孩子从现在开始学编程，他们将更容易理解这个快速变化的世界，他们将更好的融入这个世界，并且帮助我们这个快速变化的世界创造更多可能性！


### 孩子学编程两个显而易见的好处：

    编程非常有趣
    编程是一项硬核工作技能
    
#### 编程非常有趣

    技术正在成为日常生活的一部分。每家公司，慈善组织都可以从技术中受益。有应用程序可以帮助您购买，赠与，加入，玩，志愿者，连接，分享 - 你可以想象的任何东西。
    
    你的孩子曾想过创作属于他们自己的电脑游戏吗？创造自己的手机应用？他们可以在电脑上通过编程将这个想法变为现实。
    
    当孩子们开始编程时，他们会积极参与在技术方面的工作——他们不仅仅是在玩乐，而是在制作有趣的事！
    
#### 编程是一项硬核工作技能

    编码是21世纪的技能。今天的工作比以往任何时候都需要更多解决问题的能力，而且要求越来越高。
    
    职业生涯将技术作为一项不可或缺的要求。美国劳工统计局预测，在未来5年，将创造超过800多万个技术岗位。
    
    在十大增长最快的10个职业中，其中有七个是计算机科学或信息技术（IT）领域，并且这些职业已经不需要硕士或博士学位。
    
    马克扎克伯格是一名在宿舍工作的大学生，2004年他在宿舍的房间里开发了Facebook的第一个版本。10年后，每月有13.9亿人使用Facebook。历史上从来没有发生过，有一个想法，产品或服务在不到十年的时间，能够吸引数十亿的人使用。Facebook展示了技术领域的力量，比以往任何时候都更快，更快地覆盖更多人。
